```python
#fig=plt.figure()
#ax = fig.add_subplot(111)

z=result['jog_szama']
y1=result['hofer']
y2=result['bellen']
logA = np.log(z) #no need for list comprehension since all z values >= 1
logB = np.log(y1)
logC = np.log(y2)

#hofer
m1, c1 = np.polyfit(logA, logB, 1) # fit log(y) = m*log(x) + c
y_fit1 = np.exp(m1*logA + c1) # calculate the fitted values of y 

#bellen
m2, c2 = np.polyfit(logA, logC, 1) # fit log(y) = m*log(x) + c
y_fit2 = np.exp(m2*logA + c2) # calculate the fitted values of y 

#plt.plot(z, y,'.')
#plt.plot(z, y_fit2, '-')

#plt.grid(True)
#ax.set_yscale('symlog')
#ax.set_xscale('symlog')
#slope, intercept = np.polyfit(logA, logB, 1)
#plt.xlabel("Pre_referer")
#plt.ylabel("Popularity")
#ax.set_title('Pre Referral URL Popularity distribution')
#plt.show()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(np.log10(result['jog_szama']),np.log10(result['hofer']))
line = slope*result['jog_szama']+intercept

slope1, intercept1, r_value1, p_value1, std_err1 = stats.linregress(np.log(result['jog_szama']),np.log(result['bellen']))
line1 = slope1*result['jog_szama']+intercept1

# Create a trace
trace = go.Scattergl(
        x = result['jog_szama'],
        y = result['hofer'],
        mode = 'markers',
        name = 'Hofer',
        text= result['telepules'],
        marker = dict(size= 4,
                    line= dict(width=1),
                    color= 'blue',
                    opacity= 0.6
                   ),
    hoverinfo='text'
    )
    
trace1 = go.Scattergl(
        x = result['jog_szama'],
        y = result['bellen'],
        mode = 'markers',
        name = 'Bellen',
        text= result['telepules'],
        marker = dict(size= 4,
                    line= dict(width=1),
                    color= 'orange',
                    opacity= 0.4
                   ),
    hoverinfo='text'
    )

trace2 = go.Scattergl(
                  x=z,
                  y=y_fit1,
                #x=result['jog_szama'],
                  #y=line,
                  mode='lines',
                  marker=go.Marker(color='rgb(31, 119, 180)'),
                 
                  name='Fit',
                  hoverinfo='text'
                  )


trace3 = go.Scattergl(
                  x=z,
                  y=y_fit2,
                  #x=result['jog_szama'],
                  #y=line1,
                  mode='lines',
                  marker=go.Marker(color='rgb(255, 140, 0)'),
                  
                  name='Fit',
                  hoverinfo='text'
)
               
layout = go.Layout(
    title='Austrian election',
    xaxis=dict(
        type='log',
        autorange=True,
        title='log[Inhabitant number]',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        type='log',
        autorange=True,
        title='log[Candidant votes number]',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    hovermode='closest'
)

#layout=go.Layout(title="First Plot", xaxis={'title':'x1'}, yaxis={'title':'x2'})
data=[trace,trace1,trace2,trace3]  
    # Plot and embed in ipython notebook!
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='basic-scatter')
    
    # or plot with: plot_url = py.plot(data, filename='basic-line')
   
  



```
